```python
import pandas as pd
import numpy as np
from google.cloud import bigquery
from bq_helper import BigQueryHelper
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
bq_assistant = BigQueryHelper("bigquery-public-data", "openaq")
bq_assistant.head("global_air_quality")
```

```python
#air pollution form last 24h 
query = """SELECT *
            FROM `bigquery-public-data.openaq.global_air_quality`
            WHERE TIMESTAMP_DIFF(timestamp, CURRENT_TIMESTAMP(), DAY)=0 
        """
```

```python
df = bq_assistant.query_to_pandas(query)
df.head()
```

```python
df.describe()
```

```python
#analize only for pm10
pm10 = df[df['pollutant']=='pm10']
```

```python
pm10.describe()
conditions = [
    (pm10['value'] < 10),
    (pm10['value'] < 20) & (pm10['value'] > 10),
    (pm10['value'] < 40) & (pm10['value'] > 20),
    (pm10['value'] < 50) & (pm10['value'] > 40),
    (pm10['value'] < 70) & (pm10['value'] > 50),
    (pm10['value'] < 100) & (pm10['value'] > 70),
    (pm10['value'] > 100)]
choices = [1, 2, 3, 4, 5, 6, 7]
pm10['index'] = np.select(conditions, choices, default=0)
pm10= pm10[pm10['value']>0]
```

```python
pm10.describe()
pm10['value'].hist()
```

```python
plt.figure(figsize=(24,18))
plt.suptitle('PM10 [ug/m3]')
map = Basemap(llcrnrlat=-90, urcrnrlat=90, llcrnrlon=-180,urcrnrlon=180,projection='cyl')

map.drawcoastlines()
map.drawcountries()

lat,lon = map(list(pm10['latitude']),list(pm10['longitude'])) 
label = list(pm10.iloc[:,4])
plt.scatter(lon, lat, c = label, cmap=('Spectral_r'), alpha=0.75)
plt.colorbar(orientation='horizontal')
plt.show()

                              
```

```python
plt.figure(figsize=(24,10))
plt.suptitle('Average PM10 [ug/m3]')
pm10.groupby(['country'])['value'].mean().sort_values().plot.bar(y='country', rot=0, color='blue')
```

```python
pl = df[df['country']=='PL']
```

```python
plt.figure(figsize=(24,10))
plt.suptitle('Average pollutions in Poland ')
pl.groupby(['pollutant'])['value'].mean().sort_values().plot.bar(y='pollution', rot=0, color='pink')
```

```python
pl_pm10 = pl[pl['pollutant']=='pm10']
pl_pm10.describe()
```

```python
plt.figure(figsize=(24,18))
plt.suptitle('PM10 [ug/m3]')
#Poland
map = Basemap(llcrnrlat=48.8, urcrnrlat=55.0, llcrnrlon=13.8,urcrnrlon=24.4,projection='cyl')

map.drawcoastlines()
map.drawcountries()

lat,lon = map(list(pl_pm10['latitude']),list(pl_pm10['longitude'])) 
label = list(pl_pm10['value'])
plt.scatter(lon, lat, c = label, cmap=('Spectral_r'), alpha=0.75, s=100)
plt.colorbar(orientation='horizontal')
plt.show()
```

```python
plt.figure(figsize=(24,10))
plt.suptitle('PM10 in Poland ')
pl_pm10.groupby(['city'])['value'].mean().sort_values().plot.bar(y='value', rot=90, color='pink')
```
